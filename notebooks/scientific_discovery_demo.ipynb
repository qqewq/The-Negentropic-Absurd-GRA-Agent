# Scientific discovery via GRA: topological phonon insulator

```python
import sys
sys.path.append('../src')
from negentropic_gra import NegentropicGRAAgent
import numpy as np
import matplotlib.pyplot as plt

# Hypothetical phononic crystal designs as strings
designs = [f"design_{i:03d}" for i in range(500)]

def phonon_coherence(design):
    # Fidelity cost based on topological invariant calculation
    idx = int(design.split('_')[1])
    # mock: lower cost for designs with high winding number
    winding = (idx % 20) / 20.0
    return 1.0 - winding + 0.01 * np.random.randn()

def phonon_verifier(design):
    idx = int(design.split('_')[1])
    # mock log-likelihood from simulated transmission
    transmission = 1.0 / (1 + np.exp(-0.1*(idx-250)))
    return np.log(transmission + 1e-6)

agent = NegentropicGRAAgent(designs, phonon_coherence, phonon_verifier, lambda_init=0.5)
entropy_hist = []
best_hist = []
for i in range(30):
    agent.refine(1)
    info = agent.verify("acoustic_data")
    entropy_hist.append(agent.get_entropy())
    best_hist.append(agent.get_best_idea())

plt.plot(entropy_hist)
plt.xlabel('Iteration'); plt.ylabel('Entropy (bits)')
plt.title('Entropy reduction during phononic discovery')
plt.show()
print('Final design:', agent.get_best_idea())
```
