# Generative art: negentropic music motif evolution

```python
import sys
sys.path.append('../src')
from negentropic_gra import NegentropicGRAAgent
import numpy as np
import matplotlib.pyplot as plt

motifs = [f"motif_{i}" for i in range(200)]

def aesthetic_coherence(motif):
    idx = int(motif.split('_')[1])
    # lower cost for motifs with harmonic ratio
    harm = np.sin(0.2 * idx) * 0.5 + 0.5
    return 1 - harm

def audience_response(motif):
    idx = int(motif.split('_')[1])
    # mock fMRI surprise
    surprise = np.exp(-0.5 * ((idx-100)/30)**2)
    return np.log(surprise + 1e-6)

agent = NegentropicGRAAgent(motifs, aesthetic_coherence, audience_response, lambda_init=2.0)
entropies = []
for i in range(50):
    agent.refine(1)
    agent.verify("fMRI_listeners")
    entropies.append(agent.get_entropy())

plt.plot(entropies)
plt.xlabel('Iteration'); plt.ylabel('Entropy')
plt.title('Entropy reduction towards compelling motif')
plt.show()
print('Selected motif:', agent.sample_idea(temperature=0.1))
```
